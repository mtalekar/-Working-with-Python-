---
Thematic Edits

```python
l2 = [] #or l2 = list()
mylabels = list()
for gen in movies.Genre.cat.categories:
    l2.append(movies[movies.Genre == gen].BudgetMillions)
    mylabels.append(gen)
    
    
    
sns.set_style("whitegrid")
fig, ax = plt.subplots()    #this will give single plot
fig.set_size_inches(11.7,8.27)
plt.hist(l2, bins=30, stacked=True, rwidth=1, label=mylabels)     #since we are not specifing any plot, every plot 
                                                                  #will be plotted in the last plot of the sub plot
plt.title('Movie Budget Distribution', fontsize=25, color='Maroon', fontname='Serif')
plt.ylabel('Number of Movies', fontsize=20, color='purple')
plt.xlabel('Budget', fontsize=20, color='green')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(frameon = True, fancybox = True, shadow = True, framealpha = 1, prop={'size':20})
plt.show()
```
