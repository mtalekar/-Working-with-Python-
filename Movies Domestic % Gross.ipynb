```python
movies = pd.read_csv("Section6 Homework Dataset.csv", encoding= 'unicode_escape')
```

```python
movies.columns = ['DayofWeek', 'Director', 'Genre', 'MovieTitle', 'ReleaseDate',
       'Studio', 'AdjustedGross', 'Budget', 'Gross',
       'IMDbRating', 'MovieLensRating', 'Overseas', 'OverseasPercent',
       'ProfitDollars', 'ProfitPercent', 'RuntimeMinutes', 'USProfitDollars',
       'GrossUSPercent']
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.Genre = movies.Genre.astype('category')
movies.Studio = movies.Studio.astype('category')
```

```python
movies.info()
```

```python
movies.describe()
```

```python
plt.rcParams['figure.figsize'] = 15,9
```

```python
Genre_required = { 'Genre': ['action', 'comedy', 'adventure', 'animation', 'drama'] }
print(Genre_required)
```

```python
Studios_required = { 'Studio': ['Buena Vista Studios', 'Sony', 'Universal', 'WB', 'Paramount Pictures', 'Fox'] }
print(Studios_required)
```

```python
Genre_df = pd.DataFrame(Genre_required)
Studio_df = pd.DataFrame(Studios_required)
print(Genre_df)
print(Studio_df)
```

```python
merge1 = pd.merge(left=movies, right=Genre_df)
merge1
```

```python
merge1.Genre.unique()
```

```python
merge1.head()
```

```python
merge2 = pd.merge(left=merge1, right=Studio_df)
```

```python
q1 = merge2.GrossUSPercent.quantile(0.25)
q2 = merge2.GrossUSPercent.quantile(0.75)
```

```python
a = sns.boxplot(merge2.Genre, merge2.GrossUSPercent, color = 'lightgrey', showfliers = False)
```

```python
a = sns.stripplot(x=merge2.Genre, y=merge2.GrossUSPercent, hue=merge2.Studio, size=10, alpha=0.4)
```

```python
a = sns.boxplot(merge2.Genre, merge2.GrossUSPercent, color = 'lightgrey', showfliers = False)
a = sns.stripplot(x=merge2.Genre, y=merge2.GrossUSPercent, hue=merge2.Studio, size=10, alpha=0.6)
plt.title('Domestic Gross % by Genre', fontname='Arial', fontsize=35, color='black')
plt.ylabel('Gross % US', fontsize=30, color='black')
plt.xlabel('Genre', fontsize=30, color='black')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc = 'upper right', bbox_to_anchor = (1.35,1), frameon = False, prop={'size':20})
plt.show()
```
